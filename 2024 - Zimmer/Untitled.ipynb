```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=0.1
l=1

for i in range(4):
    A = i/20
    
    x = np.linspace(0, 40, 100)
    y = (-0.001*x**2+0.04*x)+(A*np.sin(l*x))
    dy = -0.002*x+0.04+A*l*np.cos(l*x)

    fig, (ax1, ax2) = plt.subplots(1,2)
    ax1.plot(x, y)
    ax2.hist(np.arctan(dy)**2, bins=5, color='skyblue', edgecolor='black')
    ax2.set_yscale('log')
    plt.show()
    print(A)
```

```python
A=0.1
l=1

for i in range(5):
    A = i/200
    
    x = np.linspace(0, 40, 100)
    y = -0.05*np.sin(x*np.pi/40)+A*np.sin(l*x)
    dy = -0.05*np.pi/40*np.cos(x*np.pi/40)+A*l*np.cos(l*x)

    fig, (ax1, ax2) = plt.subplots(1,2)
    ax1.plot(x, y)
    ax2.hist(np.arctan(dy)**2, bins=5, color='skyblue', edgecolor='black')
    ax2.set_yscale('log')
    plt.show()
    print(A)
```
