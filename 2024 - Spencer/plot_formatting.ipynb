```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
x = np.linspace(-30,30)
y = x**2
```

```python
# Plotting using defaults
plt.figure()
plt.plot(x,y,label='x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
# Specifying larger fonts, etc. (which causes the axis labels to be cropped)
figsize = (10,8)
fontsize = 35

# Plotting using those fonts (which crops the figure)
plt.figure(figsize=(figsize))
plt.plot(x,y,label='$x^2$')
plt.xlabel('x',fontsize=fontsize)
plt.ylabel('y',fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize)
```

```python
# Specifying larger fonts, etc. (adds a margin)
figsize = (10,8)
fontsize = 35
margin = 0.2

# Plotting using those fonts (which crops the figure)
plt.figure(figsize=(figsize))
plt.plot(x,y,label='$x^2$')
plt.xlabel('x',fontsize=fontsize)
plt.ylabel('y',fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize)
plt.subplots_adjust(bottom=margin) 
plt.subplots_adjust(left=margin)
```
