This first bit is exploring quadratic forms to parameterize a vapor field above a 2d facet

```python
def getsupersat2d(xgrid,ygrid,sigma_xboundary,xreduction_at_center):
    Cx = xreduction_at_center/xmax**2
    Cy = Cx
    sigma = xgrid**2*Cx + ygrid**2*Cy + sigma_xboundary - xreduction_at_center
    return sigma
```

```python
# Laying out the facet dimensions
xmax = 50
ymax = 20
x = np.linspace(-xmax,xmax,103)
y = np.linspace(-ymax,ymax,101)
ygrid,xgrid = np.meshgrid(y,x)

# Getting the supersaturation
sigma_xboundary = 5/100
xreduction_at_center = 1/100
supersat = getsupersat2d(xgrid,ygrid,sigma_xboundary,xreduction_at_center)
```

```python
plt.figure() 
ax = plt.axes(projection='3d')
ax.plot_surface(xgrid,ygrid,supersat*100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('supersaturation (%)')

fix, ax = plt.subplots()
CS = plt.contour(xgrid,ygrid,supersat*100)
plt.xlabel('x')
plt.ylabel('y')
plt.clabel(CS, inline=1, fontsize=10)
ax.set_aspect('equal')
```
