```python
import numpy as np
import matplotlib.pyplot as plt
import diffusionstuff as ds; reload(ds)
%matplotlib notebook
```

```python
Nice = np.linspace(0,1)
Nbar = 1
Nstar = .9/(2*np.pi)
Nqll = Nbar + Nstar*np.sin(2*np.pi*Nice)
plt.figure()
plt.plot(Nice,Nqll)
plt.xlabel('$N_{ice}$')
plt.ylabel('$N_{QLL}$')
plt.grid('on')
```

```python
Ntot = Nice + Nqll
plt.figure()
plt.plot(Ntot,Nqll)
plt.xlabel('$N_{tot}$')
plt.ylabel('$N_{QLL}$')
plt.grid('on')
```

```python
niter = 2
Ntotnew = np.linspace(1.0,2.0,100)
Nqll_iterative = ds.getNliq(Ntotnew,Nstar,Nbar,niter)
Ntot_onepoint = 1.61
Nqll_onepoint = ds.getNliq(Ntot_onepoint,Nstar,Nbar,niter)
plt.figure()
plt.plot(Ntot,Nqll,Ntotnew,Nqll_iterative,'r',Ntot_onepoint,Nqll_onepoint,'or')
plt.xlabel('$N_{tot}$')
plt.ylabel('$N_{QLL}$')
plt.grid('on')
```
