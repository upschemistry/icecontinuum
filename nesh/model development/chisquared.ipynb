```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib notebook
```

```python
# Make a random variable y(x)
x = np.linspace(0,1,100)
variance = 1.0
y0 = np.random.randn(len(x))*variance # This has a variance (scatter) of one
plt.figure()
plt.plot(x,y0,'o')
plt.grid('on')
plt.ylim([-4, 4])
plt.ylabel('y')
plt.xlabel('x')
plt.title('Flat line with a variance of one')
```

```python
# Give it a little structure
yp = -np.sin(x*2*np.pi)*1.1
y = y0 + yp
plt.figure()
plt.plot(x,y,'o')
plt.grid('on')
plt.ylim([-4, 4])
plt.ylabel('y')
plt.xlabel('x')
plt.title('Sinusoid with a variance of one')
```

```python
plt.figure()
plt.plot(x,y,'o',x,-yp)
plt.grid('on')
plt.ylim([-4, 4])
plt.ylabel('y')
plt.xlabel('x')
plt.title('With a stupid sinusoid')
```

```python
# Now do the chisquared test
chi2 = (y**2/variance**2).sum()/(len(y)-2); print chi2
chi2 = ((y+yp)**2/variance**2).sum()/(len(y)-2); print chi2
```
